```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Data for plotting
results_data = {
    "Method": ["Greedy", "Hill-Climb", "Substitution", "Greedy", "Hill-Climb", "Substitution", "Greedy", "Hill-Climb", "Substitution"],
    "File": ["./Entrada_1.txt", "./Entrada_1.txt", "./Entrada_1.txt",
             "./Entrada_2.txt", "./Entrada_2.txt", "./Entrada_2.txt",
             "./Entrada_3.txt", "./Entrada_3.txt", "./Entrada_3.txt"],
    "Value": [380, 417, 470, 448, 600, 382, 546, 532, 580]
}

# Creating DataFrame
results_df = pd.DataFrame(results_data)

# Convert file names to numerical x-axis for trend lines
file_map = {"./Entrada_1.txt": 1, "./Entrada_2.txt": 2, "./Entrada_3.txt": 3}
results_df['FileNum'] = results_df['File'].map(file_map)

# Define colors for each method for clarity
colors = {"Greedy": "blue", "Hill-Climb": "green", "Substitution": "orange"}

# Plotting the results
plt.figure(figsize=(10, 6))

for label, df in results_df.groupby("Method"):
    plt.scatter(df["FileNum"], df["Value"], color=colors[label], label=label, marker='o')
    # Calculate the trend line
    z = np.polyfit(df["FileNum"], df["Value"], 1)
    p = np.poly1d(z)
    plt.plot(df["FileNum"], p(df["FileNum"]), "--", color=colors[label], label=f"{label} Trend")

plt.title("Best Values by Method and File with Trend Lines")
plt.xlabel("File")
plt.ylabel("Value")
plt.xticks(ticks=[1, 2, 3], labels=["./Entrada_1.txt", "./Entrada_2.txt", "./Entrada_3.txt"])
plt.legend(title="Method")
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Correctly organized data
data = {
    "Method": ["Greedy"] * 30 + ["Substitution"] * 30 + ["Hill-Climb"] * 30,
    "File": ["./Entrada_1.txt"] * 10 + ["./Entrada_2.txt"] * 10 + ["./Entrada_3.txt"] * 10 +
            ["./Entrada_1.txt"] * 10 + ["./Entrada_2.txt"] * 10 + ["./Entrada_3.txt"] * 10 +
            ["./Entrada_1.txt"] * 10 + ["./Entrada_2.txt"] * 10 + ["./Entrada_3.txt"] * 10,
    "Value": [
        245, 318, 417, 291, 286, 375, 282, 273, 199, 350,  # Greedy Entrada 1
        336, 239, 301, 249, 208, 216, 448, 174, 279, 284,  # Greedy Entrada 2
        463, 485, 379, 414, 546, 401, 512, 458, 506, 453,  # Greedy Entrada 3
        470, 371, 398, 342, 291, 234, 275, 295, 299, 249,  # Substitution Entrada 1
        371, 371, 347, 309, 382, 376, 246, 355, 176, 325,  # Substitution Entrada 2
        533, 453, 529, 410, 496, 495, 580, 406, 462, 450,  # Substitution Entrada 3
        331, 253, 357, 254, 373, 319, 317, 373, 300, -1,   # Hill-Climb Entrada 1
        312, -1, -1, 299, -1, 328, 154, 243, -1, 154,      # Hill-Climb Entrada 2
        481, 527, 494, 452, 467, -1, 532, -1, -1, 489       # Hill-Climb Entrada 3
    ]
}

# Create DataFrame
df = pd.DataFrame(data)

# Compute the averages
average_results = df.groupby(["Method", "File"]).mean().reset_index()

# Plotting
plt.figure(figsize=(10, 6))
for label, df_group in average_results.groupby("Method"):
    plt.plot(df_group["File"], df_group["Value"], label=label, marker='o')

plt.title("Average Values by Method and File")
plt.xlabel("File")
plt.ylabel("Average Value")
plt.legend(title="Method")
plt.grid(True)
plt.show()
```

- Com esses valores podemos ver que o metodos de Substituition se torna o melhor metodos na media nas 10 execuções feitas, porem a Hill-Climb teve o problema de retorna resultados invalidos as vezes devido a sua configuração de vizinhos as vezes sobressair o limite maximo da Mochila, a possivel explicação do bom resultado do Substituition é ele ser um algoritmo de melhoria incremental onde ele tenta melhorar um resultado pré-obtido porem ele pode ser afetado pela ordenação da lista inicial de itens, já o greedy ele é um algoritmo 100% baseado na ordenação dos dados porem a randomização de ordens dele pode fazer ele se tornar favoravel.
